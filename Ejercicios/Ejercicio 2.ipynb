# Cálculo del número Pi por un método Montecarlo

Si generamos un par de números aleatorios $(x,y)$ (con probabilidad uniforme) en el segmento del plano $XY$ definido por $[-1,1]\times [-1,1]$. ¿Cuál es la probabilidad de que este punto $(x,y)$ se encuentre a una distancia menor o igual que $1$ del punto $(0,0)$?

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 100)
x = np.cos(theta)
y = np.sin(theta)
plt.figure(figsize=(4, 4))
plt.plot(x, y, 'b-')
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.title('Círculo Unitario')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```
