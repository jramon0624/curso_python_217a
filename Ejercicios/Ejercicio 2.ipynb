# Cálculo del número Pi por un método Montecarlo

Si generamos un par de números aleatorios $(x,y)$ (con probabilidad uniforme) en el segmento del plano $XY$ definido por $[-1,1]\times [-1,1]$. ¿Cuál es la probabilidad de que este punto $(x,y)$ se encuentre a una distancia menor o igual que $1$ del punto $(0,0)$?

```python
# Circulo unitario
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 100)
x = np.cos(theta)
y = np.sin(theta)
plt.figure(figsize=(4, 4))
plt.plot(x, y, 'b-')
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.title('Círculo Unitario')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
# plt.show()
```

```python
import math
import random
N = 5000 # Total de puntos a generar
n_d = 0 # Contador de puntos dentro del círculo
n = 0

dentro = []#
fuera = []#

N_list = []#
Pi_list = []#

for n in range(N):
    n+= 1
    x = random.uniform(-1,1)
    y = random.uniform(-1,1)
    N_list.append(n)#
    distancia = math.sqrt(x**2 + y**2)

    Pi_list.append(4*n_d/n)#
    if distancia <= 1:
        n_d += 1
        
        dentro.append([x, y])#
    else:#
        fuera.append([x, y])#

probabilidad = n_d / N
pi_calculado = 4 * probabilidad
print(f"Total de puntos: {N}")
print(f"Puntos dentro del círculo: {n_d}")
print(f"Probabilidad: {probabilidad}")
print(f"Estimación de pi: {pi_calculado}")
```
