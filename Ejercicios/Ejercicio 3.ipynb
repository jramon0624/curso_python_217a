# Sistema de Registro de Elementos de la Tabla Periódica
Desarrolla un programa en Python que permita registrar y consultar elementos de la tabla periódica. El programa debe usar una lista de diccionarios donde cada diccionario represente un elemento químico con sus propiedades.
## Requisitos:

El programa debe inicializarse con al menos 3 elementos químicos predefinidos en la lista.
Cada elemento debe almacenarse como un diccionario con las siguientes propiedades:

 - Símbolo químico
 - Nombre del elemento
 - Número atómico
 - Masa atómica
 - Estado físico a temperatura ambiente (sólido, líquido, gaseoso)


El programa debe mostrar un menú interactivo con las siguientes opciones:

 - 1: Agregar un nuevo elemento químico
 - 2: Listar todos los elementos disponibles
 - 3: Visualizar información detallada de un elemento específico
 - 4: Salir del programa


Al seleccionar la opción 1, el programa debe solicitar al usuario ingresar los datos del nuevo elemento y añadirlo a la lista.

Al seleccionar la opción 2, el programa debe mostrar un listado con el símbolo, nombre y número atómico de todos los elementos registrados.

Al seleccionar la opción 3, el programa debe pedir al usuario el símbolo del elemento que desea consultar y mostrar toda la información disponible sobre dicho elemento. Si el elemento no existe, mostrar un mensaje apropiado.

El programa debe continuar ejecutándose hasta que el usuario seleccione la opción para salir.

## Ejemplo de interacción:
```
=== SISTEMA DE REGISTRO DE ELEMENTOS QUÍMICOS ===
1. Agregar nuevo elemento
2. Listar elementos disponibles
3. Visualizar información de un elemento
4. Salir

Seleccione una opción: 2

ELEMENTOS DISPONIBLES:
- H (Hidrógeno) - Número atómico: 1
- O (Oxígeno) - Número atómico: 8
- Fe (Hierro) - Número atómico: 26

=== SISTEMA DE REGISTRO DE ELEMENTOS QUÍMICOS ===
1. Agregar nuevo elemento
2. Listar elementos disponibles
3. Visualizar información de un elemento
4. Salir

Seleccione una opción: 3

Ingrese el símbolo del elemento: Fe

INFORMACIÓN DEL ELEMENTO:
Símbolo: Fe
Nombre: Hierro
Número atómico: 26
Masa atómica: 55.845
Estado físico: Sólido
```
